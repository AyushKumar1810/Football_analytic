# Matplotlib, Seaborn, and mplsoccer

```python
# Let's start by creating some data
import numpy as np
import pandas as pd

# Create a time series
time = pd.date_range('2021-01-01', periods=100)
data = np.random.randn(100).cumsum()

# Create a dataframe
df = pd.DataFrame(data, index=time, columns=['value'])
df.head()
```

```python
# Matplotlib
import matplotlib.pyplot as plt

# Plot the data
plt.plot(df.index, df['value'])
```

```python
# or we can do a scatter plot
plt.scatter(df.index, df['value'])
```

```python
# Seaborn
import seaborn as sns

# Plot the data
sns.lineplot(data=df, x=df.index, y='value')
```

```python
# or we can do a scatter plot
sns.scatterplot(data=df, x=df.index, y='value')
```
