# Plotting a Pitch

The `mplsoccer` package is a great tool for plotting football pitches. 
It makes it so we can plot a pitch in just a couple of lines of code.

Throughout this course, we will be using the `mplsoccer` package to plot our pitches.

```python
from mplsoccer import Pitch
```

```python
pitch = Pitch()

fig, ax = pitch.draw()
```

```python
# We can start to change different parameters of the pitch
pitch = Pitch(
    pitch_color='grass', 
    line_color='white', 
    stripe=True, 
    linewidth=4
)
fig, ax = pitch.draw()
```

```python
# We can use a different style of pitch
# This essentially changes the dimensions of a pitch to match the data you are using
pitch = Pitch(pitch_type='statsbomb')
fig, ax = pitch.draw()
```

```python
# We can make a half pitch
pitch = Pitch(pitch_type='statsbomb', half=True)
fig, ax = pitch.draw()
```

```python
# We can also change the orientation of the pitch to be vertical
from mplsoccer import VerticalPitch
pitch = VerticalPitch(pitch_type='statsbomb')
fig, ax = pitch.draw()
```

Overall, the `mplsoccer` package is a great tool for plotting pitches and we will be using it throughout this course.
There are plenty of other parameters that can be changed, but for this course we will mostly be staying simple and using a horizontal, statsbomb pitch.
