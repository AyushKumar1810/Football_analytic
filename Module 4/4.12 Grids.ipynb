# Grids

A grid is what we can use to plot multiple graphs in a single figure. 

There are plenty of ways to create grids in Python, but I've found that the best way to do it is either by using:

1. `matplotlib`'s `gridspec` module
- This is useful when you want to create uniform grids
2. Customizing the subplots by adding axes to the figure with the location of the subplots
- This is useful when you want to create non-uniform grids



```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

import numpy as np
```

```python
x = np.array([1, 2, 3, 4])
y = np.array([10, 20, 25, 30])
```

```python
# Set up the figure and grid layout
fig = plt.figure(constrained_layout=True, figsize=(5, 5))
gs = gridspec.GridSpec(2, 2, figure=fig)

# Top row for the team names and score
ax1 = fig.add_subplot(gs[0, 0])
ax1.text(0.5, 0.5, '1', fontsize=20, ha='center')

ax2 = fig.add_subplot(gs[0, 1])
ax2.text(0.5, 0.5, '2', fontsize=20, ha='center')

ax3 = fig.add_subplot(gs[1, 0])
ax3.text(0.5, 0.5, '3', fontsize=20, ha='center')

ax4 = fig.add_subplot(gs[1, 1])
ax4.text(0.5, 0.5, '4', fontsize=20, ha='center')

# Now let's plot the data on the last subplot
ax4.set_xlabel('x')
ax4.set_ylabel('y')
ax4.plot(x, y)
```

### Now let's create a non-uniform grid using the `add_axes` method

The upside of you have way more control over subplots and their locations, spacing etc.

The downside is that you have to do a lot of manual work to get the subplots to look good.

```python
fig = plt.figure(figsize=(5, 5))

# Create the first subplot
# The first two arguments are the location of the bottom left corner of the subplot
# The last two arguments are the width and height of the subplot

ax1 = fig.add_axes([0, 0, 0.25, 0.25])

# Create the second subplot
ax2 = fig.add_axes([0.5, 0, 0.25, 0.25])

# Create the third subplot
ax3 = fig.add_axes([0, 0.5, 0.25, 0.25])

# Create the fourth subplot
ax4 = fig.add_axes([0.5, 0.5, 0.25, 0.25])
```
