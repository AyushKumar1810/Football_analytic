# Correlation

The correlation coefficient is a measure of linear association between two variables. It ranges from -1 to 1. A value of 1 indicates a perfect positive linear relationship, and a value of -1 indicates a perfect negative linear relationship. A value of 0 indicates no linear relationship between the two variables.

Correlation does not imply causation. It only measures the strength of the linear relationship between two variables.

Correlation is important in football analytics because it helps us understand the relationship between different variables. For example, we can calculate the correlation between the number of goals scored and the final position in the league table. We can also calculate the correlation between the number of passes completed and the number of goals scored to understand the relationship between these two variables.

In python we can use the `corr` method to calculate the correlation between two variables.

```python
import pandas as pd
```

```python
# Importing the dataset
data = pd.read_csv('~/Documents/GitHub/complete-football-analytics/Module 5/big5-clean.csv')
```

```python
data.columns
```

```python
# Calculating the correlation between two variables
data['Gls'].corr(data['PrgC'])
```

```python
# Calculating the correlation between all the variables
# These variables must be numerical
data[['Gls', 'Ast', 'PK', 'xG', 'PrgC', 'PrgP', 'PrgR']].corr()
```

```python
# Plotting the correlation matrix
import seaborn as sns
sns.heatmap(data[['Gls', 'Ast', 'PK', 'xG', 'PrgC', 'PrgP', 'PrgR']].corr(), annot=True)
```
